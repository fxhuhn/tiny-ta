# Applying the RSL function to the S&P 500 index 

```python
# import pandas as pd
import yfinance as yf
from tiny_ta import rsl
```

```python
# Download the S&P 500 index price data from Yahoo
sp_500 = yf.download("^GSPC", interval="1wk")
```

```python
sp_500["rsl"] = rsl(sp_500.Close)
```

```python
sp_500["buy"] = sp_500[sp_500.rsl.shift(1) > 1]["Close"]
```

```python
sp_500["buy_diff"] = sp_500.buy.diff()
sp_500["close_diff"] = sp_500.Close.diff()
```

```python
sp_500["2023-01-01":"2024-01-01"]
```

```python
sp_500["2023-01-01":"2024-01-01"].close_diff.sum()
```

```python
sp_500["2023-01-01":"2024-01-01"].buy_diff.sum()
```

```python
sp_500[["Close", "buy", "rsl"]]["2023-01-01":"2024-01-01"].plot(
    subplots=True, figsize=(16, 9), title="S&P 500"
)
```
